```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_moons

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
# Classification Dataset
X, y = make_moons(100, noise=0.25, random_state=2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
model1 = Sequential()

model1.add(Dense(128, activation = 'relu', input_dim = 2))
model1.add(Dense(128, activation = 'relu'))

model1.add(Dense(1, activation = 'sigmoid'))

model1.summary()
```

```python
# ziyada epoches khud rakhain han ta k overfitting ho  
adam = Adam(learning_rate=0.01)
model1.compile(loss = 'binary_crossentropy', optimizer = adam, metrics=['accuracy'])

history = model1.fit(X,y, epochs=2000, validation_split=0.2, verbose=1 )
```

```python
plot_decision_regions(X, y.astype('int'), clf=model1, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()
```

```python
# model with regularization
model2 = Sequential()

model2.add(Dense(128, activation = 'relu', input_dim = 2, kernel_regularizer = tf.keras.regularizers.l2(0.03)))
model2.add(Dense(128, activation = 'relu', kernel_regularizer = tf.keras.regularizers.l2(0.03)))

model2.add(Dense(1, activation = 'sigmoid'))

model2.summary()
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss = 'binary_crossentropy', optimizer = adam, metrics=['accuracy'])

history2 = model2.fit(X,y, epochs=2000, validation_split=0.2, verbose=1 )
```

```python
plot_decision_regions(X, y.astype('int'), clf=model2, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
# by regularization weights decay kr gay (goes to zero ho gay)
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
plt.show()
```

```python
model1_weight_layer1 = model1.get_weights()[0].reshape(256)
model2_weight_layer1 = model2.get_weights()[0].reshape(256)
```
