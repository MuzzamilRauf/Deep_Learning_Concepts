```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

Genetrate Data

```python
X_train = np.linspace(-1, 1, 20)
```

```python
y_train = np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  , 
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475, 
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,  
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])
```

```python
X_test = np.linspace(-1, 1, 20)
```

```python
y_test = np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,  
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X_train, y_train, c='red', label='Train')
plt.scatter(X_test, y_test, c='blue', label='Test')
plt.legend()
plt.show()
```

Regression Model

```python
model = Sequential()

model.add(Dense(128,activation ='relu',input_dim = 1))
model.add(Dense(128,activation ='relu'))

model.add(Dense(1,activation = 'linear'))

adam = Adam(learning_rate = 0.01)

model.compile(loss = 'mse', optimizer = 'adam', metrics=['mse'])

history = model.fit(X_train, y_train, epochs=500, 
                    validation_data = (X_test, y_test),
                    verbose=False)
```

```python
history.history
```

```python
# evaluate the model
_, train_mse = model.evaluate(X_train, y_train, verbose=0)
_, test_mse = model.evaluate(X_test, y_test, verbose=0)
print('Train: {}, Test: {}'.format(train_mse, test_mse))
```

```python
y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X_train,y_train,c='red', label = 'Training')
plt.scatter(X_test,y_test,c='blue', label = 'Testinig')
plt.plot(X_test,y_pred)
plt.ylim((-1.5, 1.5))
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()
```

```python
# this is overfitting how it can solved. By using Dropouts.

model1 = Sequential()

model1.add(Dense(128,activation ='relu',input_dim = 1))
model1.add(Dropout(0.5))
model1.add(Dense(128,activation ='relu'))
model1.add(Dropout(0.5))
model1.add(Dense(1,activation = 'linear'))

adam = Adam(learning_rate = 0.01)

model1.compile(loss = 'mse', optimizer = 'adam', metrics=['mse'])

history = model1.fit(X_train, y_train, epochs=500, 
                    validation_data = (X_test, y_test),
                    verbose=False)
```

```python
# evaluate the model
_, train_mse1 = model1.evaluate(X_train, y_train, verbose=0)
_, test_mse1 = model1.evaluate(X_test, y_test, verbose=0)
print('Train: {}, Test: {}'.format(train_mse1, test_mse1))
```

```python
y_pred1 = model.predict(X_test)
```
